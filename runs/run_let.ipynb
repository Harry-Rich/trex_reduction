```python
import mcstasscript as ms
import sys
import os


parent = os.path.dirname(os.getcwd())
sys.path.append(os.path.join(parent, "instruments"))


import make_LET
from samples  import add_vanadium_sample, add_benzene_sample

from inst_settings import benz_kwargs, let_kwargs
```

```python
inp = os.path.join(os.getcwd(), "LET")
os.makedirs(inp, exist_ok=True)

LET_benz = make_LET.make(name = 'LET_benz', input_path=inp)
add_benzene_sample(LET_benz,**let_kwargs, **benz_kwargs)

LET_vanad = make_LET.make(name = 'LET_vanad', input_path=inp)
add_vanadium_sample(LET_vanad,**let_kwargs)

LET_vanad.settings(ncount=ncount, NeXus=True,mpi=mpi, suppress_output=suppress_output)
LET_benz.settings(ncount=ncount, NeXus=True,mpi=mpi, suppress_output=suppress_output)


LET_benz.backengine()
LET_vanad.backengine()
```
